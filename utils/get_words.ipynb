```python
import epub2txt
from collections import defaultdict

res = epub2txt.epub2txt('../books/GoT.epub')

res = res.replace('’', "'")
res = res.replace('“', '"')
res = res.replace('”', '"')

new_res = ""

for line in res.splitlines():
    if len(line) > 0 and line[0] != '@':
        new_res += line + '\n'

res = new_res
```

```python
from nltk.tokenize import TweetTokenizer
from nltk.tag import pos_tag

tknzr = TweetTokenizer()

pos = pos_tag(tknzr.tokenize(res))
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

based_words = []

for word, tag in pos:
    base_word = word

    if tag.startswith('NN'):
        base_word = lemmatizer.lemmatize(word, 'n')
    if tag.startswith('VB'):
        base_word = lemmatizer.lemmatize(word, 'v')
    if tag.startswith('JJ'):
        base_word = lemmatizer.lemmatize(word, 'a')
    if tag.startswith('RB'):
        base_word = lemmatizer.lemmatize(word, 'r')
    
    if tag not in '.,' and word not in '";…—-:{}[]-\'‘':
        based_words.append(base_word.lower() if base_word != 'I' else base_word)


print(based_words[1000:1050])
```

```python
from nltk.probability import FreqDist

freq_dist = FreqDist(based_words)

print(freq_dist.most_common())
```
